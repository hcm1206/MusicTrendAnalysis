```python
import pandas as pd

# Load the datasets
circle_chart_df = pd.read_csv('circle_chart_removed_na.csv')
finish_df = pd.read_csv('finish.csv')

# Merge the datasets based on the condition: date == Date and New_Artist == Artist
merged_df = pd.merge(circle_chart_df, finish_df, left_on=['date', 'New_Artist'], right_on=['Date', 'Artist'], how='inner')

merged_df.drop(['Artist_y','Date'], axis=1, inplace=True)

# Save the merged dataset to a new file
merged_file_path = 'merged_circle_finish.csv'
merged_df.to_csv(merged_file_path, index=False)

merged_file_path
```
