```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 데이터 준비 (data는 사전에 준비되어야 함)
# X에는 과거 피처들 (예: 날짜, 아티스트 정보, 장르 등), y에는 Rank
X = data.drop(['Rank', 'date', 'New_Artist', 'Title', 'Production', 'Genre', 'Genre_split', 'season'], axis=1) 
y = data["Rank"]  # 랭킹 데이터를 목표 변수로 설정

# 데이터 전처리
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 훈련용 데이터와 테스트 데이터 나누기
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# XGBoost 모델 설정 및 학습
xgb_model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5, n_jobs=-1, random_state=42)
xgb_model.fit(X_train, y_train)

# 예측
y_pred = xgb_model.predict(X_test)

# 성능 평가
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

# 결과 출력
print(f"RMSE: {rmse}")
print(f"R²: {r2}")
```

```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 날짜 형식을 datetime으로 변환하여 필터링
data['date'] = pd.to_datetime(data['date'])  # 날짜 데이터를 datetime 형식으로 변환

# 9월 8일 데이터를 필터링
sept_8_data = data[data['date'] == pd.Timestamp('2024-09-08')]
data_except_sept_8 = data[data['date'] != pd.Timestamp('2024-09-08')]

# 9월 8일 데이터가 비어있는지 확인
print(f"September 8 data size: {sept_8_data.shape}")

# X에는 과거 피처들 (예: 날짜, 아티스트 정보, 장르 등), y에는 Rank
X = data_except_sept_8.drop(['Rank', 'date', 'New_Artist', 'Title', 'Production', 'Genre', 'Genre_split', 'season'], axis=1)
y = data_except_sept_8["Rank"]

# 9월 8일 검증용 데이터
X_sept_8 = sept_8_data.drop(['Rank', 'date', 'New_Artist', 'Title', 'Production', 'Genre', 'Genre_split', 'season'], axis=1)
y_sept_8 = sept_8_data['Rank']

# 데이터 전처리
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_sept_8_scaled = scaler.transform(X_sept_8)  # 9월 8일 데이터도 동일하게 스케일링

# 훈련용 데이터와 테스트 데이터 나누기
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

### 랜덤포레스트 모델
rf_model = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)
rf_model.fit(X_train, y_train)

# 랜덤포레스트 예측
y_pred_rf = rf_model.predict(X_test)
y_pred_rf_sept_8 = rf_model.predict(X_sept_8_scaled)

# 랜덤포레스트 성능 평가
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
r2_rf = r2_score(y_test, y_pred_rf)

# 9월 8일에 대한 랜덤포레스트 성능 평가
rmse_rf_sept_8 = np.sqrt(mean_squared_error(y_sept_8, y_pred_rf_sept_8))
r2_rf_sept_8 = r2_score(y_sept_8, y_pred_rf_sept_8)

### XGBoost 모델
xgb_model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5, n_jobs=-1, random_state=42)
xgb_model.fit(X_train, y_train)

# XGBoost 예측
y_pred_xgb = xgb_model.predict(X_test)
y_pred_xgb_sept_8 = xgb_model.predict(X_sept_8_scaled)

# XGBoost 성능 평가
rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
r2_xgb = r2_score(y_test, y_pred_xgb)

# 9월 8일에 대한 XGBoost 성능 평가
rmse_xgb_sept_8 = np.sqrt(mean_squared_error(y_sept_8, y_pred_xgb_sept_8))
r2_xgb_sept_8 = r2_score(y_sept_8, y_pred_xgb_sept_8)

# 결과 출력
print("### 랜덤포레스트 성능 ###")
print(f"Test RMSE (RandomForest): {rmse_rf}")
print(f"Test R² (RandomForest): {r2_rf}")
print(f"September 8 RMSE (RandomForest): {rmse_rf_sept_8}")
print(f"September 8 R² (RandomForest): {r2_rf_sept_8}")

print("\n### XGBoost 성능 ###")
print(f"Test RMSE (XGBoost): {rmse_xgb}")
print(f"Test R² (XGBoost): {r2_xgb}")
print(f"September 8 RMSE (XGBoost): {rmse_xgb_sept_8}")
print(f"September 8 R² (XGBoost): {r2_xgb_sept_8}")
```
