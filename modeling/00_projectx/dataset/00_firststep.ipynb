### 계절 데이터값과 원주율 일치 0 ~ 2pi 계산한 값 (1달 31일 1년 372일 기준) 
### 새로운컬럼 'date_pi' 생성

```python
import pandas as pd
import numpy as np



# 1년을 372일로 고정 (한 달 = 31일 기준)
total_days_in_year = 372

# 날짜를 각도로 변환하는 함수 (각도는 0부터 2 * pi까지, 372일을 기준으로 함)
def date_to_angle(date):
    # 날짜의 월과 일을 기반으로 372일 중 몇 번째 날인지 계산
    month = date.month
    day = date.day
    day_of_year = (month - 1) * 31 + day  # 1년을 31일로 고정했을 때의 날 수 계산
    angle = 2 * np.pi * day_of_year / total_days_in_year  # 각도 변환
    return angle

# 변환된 결과를 적용 (하나의 컬럼 'angle'에 각도 값을 저장)
df['date_pi'] = df['date'].apply(date_to_angle)

# 결과 확인
df[['date', 'date_pi']].head()
```

```python
import matplotlib.pyplot as plt

# 1년을 372일로 고정 (한 달 = 31일 기준)
total_days_in_year = 372

# 날짜를 각도로 변환하는 함수 (각도는 0부터 2 * pi까지, 372일을 기준으로 함)
def date_to_angle(date):
    month = date.month
    day = date.day
    day_of_year = (month - 1) * 31 + day  # 1년을 31일로 고정했을 때의 날 수 계산
    angle = 2 * np.pi * day_of_year / total_days_in_year  # 각도 변환
    return angle

# 변환된 결과를 적용 (하나의 컬럼 'angle'에 각도 값을 저장)
df['season_angle'] = df['date'].apply(date_to_angle)

# 계절을 결정하는 함수
def determine_season(month):
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Autumn'
    else:
        return 'Winter'

# 각 날짜에 해당하는 계절을 추가
df['season'] = df['date'].apply(lambda x: determine_season(x.month))

# 그래프 그리기
plt.figure(figsize=(10, 6))

# 각 계절별로 다른 색으로 시각화
for season in ['Spring', 'Summer', 'Autumn', 'Winter']:
    season_data = df[df['season'] == season]
    plt.scatter(season_data['date'], season_data['season_angle'], label=season)

plt.title('Seasonal Representation Using Angle')
plt.xlabel('Date')
plt.ylabel('Angle (radians)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd

# 계절별 고정된 각도를 라디안 값으로 설정 (0 ~ 2π 사이)
season_start_angles = {
    'Spring': 0,                # 봄: 0 라디안 (0도)
    'Summer': np.pi / 2,         # 여름: π/2 라디안 (90도)
    'Autumn': np.pi,             # 가을: π 라디안 (180도)
    'Winter': 3 * np.pi / 2      # 겨울: 3π/2 라디안 (270도)
}

# 계절을 결정하는 함수
def determine_season(month):
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Autumn'
    else:
        return 'Winter'

# 데이터의 'date' 컬럼이 datetime 형식인지 확인 후 변환
if not pd.api.types.is_datetime64_any_dtype(df['date']):
    df['date'] = pd.to_datetime(df['date'])

# 각 날짜에 해당하는 계절을 추가
df['season'] = df['date'].apply(lambda x: determine_season(x.month))

# 계절에 따른 고정된 라디안 각도를 추가
df['season_pi'] = df['season'].map(season_start_angles)
```

### 컬럼 'date', 'Title'을 기준으로 차트에 진입하면 +1 count 하여
### 새로운컬럼 'weekly_count'와 weekly_count의 max값인 'cumulative_count'  컬럼 생성

```python
# Title별로 날짜가 오래된 순서대로 weekly_count를 증가시키는 방식으로 수정
df = df.sort_values('date')

# Title과 date를 기준으로 오래된 순서부터 weekly_count 증가
df['weekly_count'] = df.groupby('Title')['date'].rank(method='dense').astype(int)

# cumulative_count는 Title과 date를 기준으로 weekly_count의 최대 값을 계산
df['cumulative_count'] = df.groupby(['Title', 'date'])['weekly_count'].transform('max')

# 결과 확인
max_weekly_count = df['weekly_count'].max()
max_weekly_count

```
