```python
data2['Genre_split'] = data2['Genre_split'].apply(lambda x: [genre_dict.get(genre, genre) for genre in x])
```

```python
def convert_runtime_to_seconds(runtime):
    try:
        minutes, seconds = map(int, runtime.split(':'))
        return minutes * 60 + seconds
    except:
        return None

data2['Runtime'] = data2['Runtime'].apply(convert_runtime_to_seconds)
```

```python
# 'ky_chart_on' 열을 int 형으로 변환 (True -> 1, False -> 0)
data2['ky_chart_on'] = data2['ky_chart_on'].astype(int)

# 변환 후 결과 확인
data2.head().T
```

```python
import pickle
from sklearn.preprocessing import StandardScaler

# Runtime 컬럼에 스케일러 적용
runtime_scaler = StandardScaler()
data2['Runtime'] = runtime_scaler.fit_transform(data2[['Runtime']])

# WeeklyViews 컬럼에 스케일러 적용
views_scaler = StandardScaler()
data2['Weekly Views'] = views_scaler.fit_transform(data2[['Weekly Views']])

# 각각의 스케일러를 피클로 저장
with open('runtime_scaler.pkl', 'wb') as f:
    pickle.dump(runtime_scaler, f)

with open('views_scaler.pkl', 'wb') as f:
    pickle.dump(views_scaler, f)

# 피클로부터 스케일러 불러오기
with open('runtime_scaler.pkl', 'rb') as f:
    loaded_runtime_scaler = pickle.load(f)

with open('views_scaler.pkl', 'rb') as f:
    loaded_views_scaler = pickle.load(f)

# 불러온 스케일러로 데이터를 변환
data2['Runtime'] = loaded_runtime_scaler.transform(data2[['Runtime']])
data2['Weekly Views'] = loaded_views_scaler.transform(data2[['Weekly Views']])
```

```python
import pickle
from sklearn.preprocessing import MinMaxScaler

# Min-Max Scaler 적용
n_score_scaler = MinMaxScaler()
g_score_scaler = MinMaxScaler()

# n_score에 Min-Max 스케일링 적용
data2['n_score'] = n_score_scaler.fit_transform(data2[['n_score']])

# g_score에 Min-Max 스케일링 적용
data2['g_score'] = g_score_scaler.fit_transform(data2[['g_score']])

# 각각의 스케일러를 피클로 저장
with open('n_score_scaler.pkl', 'wb') as f:
    pickle.dump(n_score_scaler, f)

with open('g_score_scaler.pkl', 'wb') as f:
    pickle.dump(g_score_scaler, f)

# 피클로부터 스케일러 불러오기
with open('n_score_scaler.pkl', 'rb') as f:
    loaded_n_score_scaler = pickle.load(f)

with open('g_score_scaler.pkl', 'rb') as f:
    loaded_g_score_scaler = pickle.load(f)

# 불러온 스케일러로 다시 데이터를 변환 (필요 시)
data2['n_score'] = loaded_n_score_scaler.transform(data2[['n_score']])
data2['g_score'] = loaded_g_score_scaler.transform(data2[['g_score']])
```

```python

def assign_season(date):
    # 날짜에서 월을 추출 (문자열의 5번째와 6번째 글자가 월을 나타냄)
    month = date[5:7]
    
    if month in ['03', '04', '05']:
        return 'spring'
    elif month in ['06', '07', '08']:
        return 'summer'
    elif month in ['09', '10', '11']:
        return 'autumn'
    else:
        return 'winter'
    
# apply를 사용해 'season' 컬럼 생성
data2['season'] = data2['date'].apply(assign_season)
```

```python
# 원핫 인코딩 적용
data2 = pd.get_dummies(data2, columns=['season'], prefix='season')
```

```python
season_columns = [col for col in data2.columns if 'season_' in col]
data2[season_columns] = data2[season_columns].astype(int)
```
