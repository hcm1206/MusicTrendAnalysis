```python
import pandas as pd

data=pd.read_csv('DBdata.csv')
# data.head()
```

```python
# Rank_lag_1, Rank_lag_2, Rank_lag_3 컬럼 드랍
data = data.drop(columns=['Rank_lag_1', 'Rank_lag_2', 'Rank_lag_3'])

# 변경된 데이터프레임 확인
data.head()
```

```python
# date 컬럼을 datetime 형식으로 변경
# data['date'] = pd.to_datetime(data['date'])

# 변경된 데이터프레임 확인
data['date'] = pd.to_datetime(data['date']).dt.floor('S')
# data.info()



```

```python
# Rank_lag_1, Rank_lag_2, Rank_lag_3 컬럼을 생성 및 초기화
data['Rank_lag_1'] = 201
data['Rank_lag_2'] = 201
data['Rank_lag_3'] = 201

# 데이터프레임을 날짜 기준으로 정렬
data = data.sort_values(by='date')

# 일주일 전, 2주일 전, 3주일 전의 Rank를 찾는 함수 정의
def find_previous_rank(row, weeks_ago):
    target_date = row['date'] - pd.Timedelta(weeks=weeks_ago)
    match = data[(data['date'] == target_date) &
               (data['Title'] == row['Title']) &
               (data['New_Artist'] == row['New_Artist'])]
    
    # 매칭된 데이터가 있으면 그 행의 Rank 값을 반환, 없으면 201 반환
    if not match.empty:
        return match['Rank'].values[0]
    else:
        return 201

# 각 행에 대해 Rank_lag_1, Rank_lag_2, Rank_lag_3 값을 계산
data['Rank_lag_1'] = data.apply(lambda row: find_previous_rank(row, 1), axis=1)
data['Rank_lag_2'] = data.apply(lambda row: find_previous_rank(row, 2), axis=1)
data['Rank_lag_3'] = data.apply(lambda row: find_previous_rank(row, 3), axis=1)

# 결과 확인
data[['date', 'Title', 'New_Artist', 'Rank', 'Rank_lag_1', 'Rank_lag_2', 'Rank_lag_3']].head()
```

```python

# New_Artist가 '악뮤'이고, Title이 'Love Lee'인 데이터를 date 기준으로 정렬 후 상위 50개 출력
data_filtered = data[(data['New_Artist'] == '임영웅') & (data['Title'] == '이제 나만 믿어요')].sort_values(by='date')

# 정렬된 데이터에서 상위 50개 확인
data_filtered.head(50)
```
