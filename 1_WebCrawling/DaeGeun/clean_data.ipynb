```python
file_path = 'circle_chart_data_with_dates.csv'
```

```python
data = pd.read_csv(file_path)
```

```python
# 컬럼 이름 출력
print(data.columns)
```

```python
# Strip any leading/trailing spaces from the column names again to clean them up fully
data.columns = data.columns.str.strip()

# Now we will try to filter the 'date' and 'Artist' columns
filtered_data = data[['date', 'Artist']]

# Display the first few rows to confirm the result
filtered_data.head()
```

```python
# 필터링된 데이터를 CSV 파일로 저장
filtered_data.to_csv('filtered_data.csv', index=False)
```

```python
# 날짜별로 데이터를 그룹화한 후, 갑 그룹에서 중복된 가수 제거
unique_data = filtered_data.groupby('date').apply(lambda x: x.drop_duplicates(subset='Artist')).reset_index(drop=True)

# 결과 확인
unique_data.head()
```

```python
# 결과를 CSV 파일로 저장
output_file = 'cleaned_artist_data.csv'  # 저장할 파일 경로
unique_data.to_csv(output_file, index=False)

print(f"중복된 가수를 제거한 파일이 '{output_file}'로 저장되었습니다.")
```
