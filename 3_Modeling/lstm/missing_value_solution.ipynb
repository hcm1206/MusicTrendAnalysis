```python
import pandas as pd

# 데이터 로드
df = pd.read_csv(r'D:\MusicTrendAnalysis\data\final_total_data.csv')
```

```python
# 본래 데이터의 총 행 개수 확인
total_rows = df.shape[0]
print(f"본래 데이터의 총 행 개수: {total_rows}")
```

```python
# 중복 제거: 같은 날짜(date)에 같은 노래(Title)를 기준으로 중복된 행을 제거
df_deduplicated = df.drop_duplicates(subset=['date', 'Title'])
```

```python
# 중복 제거 후 총 행 개수 확인
total_rows_deduplicated = df_deduplicated.shape[0]
removed_rows = total_rows - total_rows_deduplicated
print(f"중복 제거 후 총 행 개수: {total_rows_deduplicated}")
print(f"제거된 행 개수: {removed_rows}")
```

```python
# 날짜별로 Rank가 1~200 사이에 빠진 값이 있는지 확인
missing_ranks = {}

for date, group in df_deduplicated.groupby('date'):
    ranks = group['Rank'].dropna().astype(int)  # Rank 값을 정수로 변환하고 NaN 제거
    missing_for_date = set(range(1, 201)) - set(ranks)  # 1~200에서 빠진 Rank 찾기
    if missing_for_date:
        missing_ranks[date] = sorted(missing_for_date)

# 누락된 랭크 출력
for date, missing in missing_ranks.items():
    print(f"{date}에 누락된 Rank: {missing}")
```

```python
# 총 누락된 Rank의 개수 계산 및 출력
total_missing_ranks = sum(len(ranks) for ranks in missing_ranks.values())
print(f"총 누락된 Rank 개수: {total_missing_ranks}")
```
